```python
region_counts = df1.groupby('Region')['CustomerID'].count()

plt.figure(figsize=(8, 5))
region_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Number of Customers by Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Number of Customers', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
df2.columns
```

```python
region_counts = df2.groupby('Category')['Price'].count()

plt.figure(figsize=(8, 5))
region_counts.plot(kind='bar', color='pink', edgecolor='black')

plt.title('Price by Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
region_counts = df3.groupby('Quantity')['Price'].count()

plt.figure(figsize=(8, 5))
region_counts.plot(kind='bar', color='pink', edgecolor='black')

plt.title('Price by Region', fontsize=16)
plt.xlabel('Quantity', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
df1=df
df['SignupDate'] = pd.to_datetime(df['SignupDate'])

df['SignupMonth'] = df['SignupDate'].dt.month

plt.figure(figsize=(10, 6))

for region in df['Region'].unique():
    region_data = df[df['Region'] == region]['SignupMonth']
    plt.hist(region_data, bins=4, alpha=0.7, label=region)

plt.title('Customer Signups by Month (Grouped by Region)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Region')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
